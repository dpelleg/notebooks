```python
import pandas as pd
```

```python
datapath = 'data/'
dr = pd.read_csv(datapath + 'dist_driving_Haifa.csv')
tr = pd.read_csv(datapath + 'dist_transit_Haifa Hof Hacarmel.csv')
```

```python
df = pd.merge(dr, tr, on='name', suffixes=('_drive', '_transit'))
```

```python
df.columns
```

```python
df = df[['distance_drive', 'duration_drive', 'speed_drive', 'crow_dist_drive', 'name',
       'distance_transit', 'duration_transit', 'speed_transit', 'שם יישוב_drive']].rename(columns={'שם יישוב_drive' : 'שם'})
```

```python
df['ratio'] = df['duration_transit']/df['duration_drive']
df2 = df.query('duration_drive > 1800').sort_values(by='ratio', ascending=False)
```

```python
df2[['שם', 'duration_drive', 'duration_transit', 'ratio']][0:20]
```

```python
#df3 = dr[['distance', 'duration', 'speed', 'crow_dist', 'שם יישוב']]
df3 = dr.query('duration > 1800').sort_values(by='ratio')[['distance', 'duration', 'speed', 'crow_dist', 'שם יישוב']].drop(columns=['duration', 'speed'])
df3['crow_dist'] = df3['crow_dist'].astype(int)
df3.rename(columns={'distance' : 'מרחק כביש', 'crow_dist' : 'קו אוירי'})[0:20]
```
