```python
import pandas as pd
# source: page 7 of https://www.cbs.gov.il/he/mediarelease/DocLib/2023/109/15_23_109b.pdf
df = pd.read_csv('../data/gini.csv')
```

```python
df
```

```python
df.dropna(how='all', axis=1, inplace=True)
df.dropna(how='all', axis=0, inplace=True)
```

```python
df.drop([58], inplace=True)
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df['year'] = df.apply(lambda x: x['Unnamed: 44'] if (pd.isna(x['Unnamed: 42'])) else x['Unnamed: 42'], axis=1)
```

```python
df.drop(columns=['Unnamed: 44', 'Unnamed: 42'], inplace=True)
```

```python
df.rename(columns={'Unnamed: 3' : 'עליון', 'Unnamed: 32' : 'תחתון', 'Unnamed: 35' : 'סך הכל'}, inplace=True)
```

```python
df.year = df.year.astype(str)
```

```python
df.drop(columns=['סך הכל'], inplace=True)
df
```

```python
import seaborn as sns 
import re
import numpy as np
```

```python
def extract_year(s):
    if pd.isna(s):
      return ""
    return int(re.match(r".*(\d{4}).*", s).group(1))
```

```python
df.index=df['year'].apply(extract_year)
df.drop(columns='year', inplace=True)
```

```python
df
```

```python
def gini(x):
    total = 0
    for i, xi in enumerate(x[:-1], 1):
        total += np.sum(np.abs(xi - x[i:]))
    return total / (len(x)**2 * np.mean(x))
```

```python
gini_v = df.apply(lambda x: gini(np.array(x.to_list())), axis=1)
```

```python
gini_v = gini_v[~gini_v.index.duplicated(keep='first')]
```

```python
gini_v.plot(ylabel='gini', xlabel='year')
```
